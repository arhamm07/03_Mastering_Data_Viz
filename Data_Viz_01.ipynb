# DATA VISUALIZATION

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
sns.scatterplot(data=df,x='age',y='fare',hue='class')
```

```python
# Adding Fig Size
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='age',y='fare',hue='class')
```

```python
# Changing Color of the Figure
color = {'First':'#588157','Second':'#3a5a40','Third':'#344e41'}
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='age',y='fare',hue='class',palette=color)
```

```python
color = {'First':'#588157','Second':'#3a5a40','Third':'#344e41'}
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='age',y='fare',hue='class',palette=color)
#Adding labels to the chart

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age Vs Fare')
```

```python
color = {'First':'#588157','Second':'#3a5a40','Third':'#344e41'}
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='age',y='fare',hue='class',palette=color)

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age Vs Fare')

# Saving the plot
plt.savefig("./Plots/Titanic_AgeVsFare.png",dpi = 300)
```

```python
# Changing style of the plot

sns.set_style('darkgrid')

color = {'First':'#588157','Second':'#3a5a40','Third':'#344e41'}
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='age',y='fare',hue='class',size='age',palette=color)

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age Vs Fare')
```

`Interpretation of the plot:`
1. We have Age of Passengers at X-axis and Fare of Passengers at Y-axis (As Shown in Fig 1) 
2. The Scatter plot shows the relationship between age and fare of the passengers (As Shown in Fig 1)
3. We further grouped the passengers with their class with different colors (As Shown in Fig 1)
4. Two male passengers gave the highest fare in titanic and they might be travelling in first class (As Shown in Fig 1)
5. The oldest passenger of the titanic is a male of 80 years (As Shown in Fig 1)

```python
sns.set_style('darkgrid')

color = {'First':'#588157','Second':'#3a5a40','Third':'#344e41'}
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='sex',y='age',hue='class',palette=color)

plt.xlabel('Sex')
plt.ylabel('Age')
# plt.title('Scatter Plot of Age Vs Fare')
```

```python
sns.set_style('darkgrid')

color = {'First':'#588157','Second':'#3a5a40','Third':'#344e41'}
plt.figure(figsize=(10,6))
sns.boxenplot(data=df,x='sex',y='age',hue='class',palette=color)

plt.xlabel('Sex')
plt.ylabel('Age')
plt.xticks(ticks = [0,1],labels=['MALE','FEMALE'],rotation= 0)
```

```python
# PLotting using plotly

fig = px.scatter(df,x='age',y='fare',color='class')
fig.show()
```

```python
sns.set_style('darkgrid')

color = {'First':'#588157','Second':'#3a5a40','Third':'#344e41'}
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='sex',y='age',hue='class',palette=color)

plt.xlabel('Sex')
plt.ylabel('Age')
```

```python
sns.set_style('darkgrid')

color = {'First':'#588157','Second':'#3a5a40','Third':'#344e41'}
plt.figure(figsize=(10,6))
sns.violinplot(data=df,x='sex',y='age',hue='class',palette=color)

plt.xlabel('Sex')
plt.ylabel('Age')
```
